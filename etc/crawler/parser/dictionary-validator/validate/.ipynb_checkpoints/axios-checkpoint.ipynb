```python
f = open('../../../data/dictionarys/axios.pkl','rb')
axios = pickle.load(f)
```

```python
for idx,dic in enumerate(axios):
    if dic['mean'] == ['에러 발생']:
        print(dic['word'])
        mean = input()
        tmp = {'word' : dic['word'],'mean':[mean]}
        print(tmp)
        axios[idx] = tmp
```
